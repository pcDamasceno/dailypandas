```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Chandelier_Exit_2min_List_of_Trades_2024-07-09.csv")

df['Date/Time'] = pd.to_datetime(df['Date/Time'])
#FILTER BY DATE
#df = df.loc[(df['Date/Time'] >= '2024-07-01') & (df['Date/Time'] <= '2024-07-31')]


df = df[df['Signal'] != 'Buy']
df
```

```python
# Percentage of trade results
df['Signal'].value_counts(normalize=True)*100
```

```python
# Descriptive statistics for numeric columns
desc_stats = df.describe()

# Summary of trade types and signals
trade_types_summary = df['Type'].value_counts()
signals_summary = df['Signal'].value_counts()

# Analyzing overall profitability
total_profit_brl = df['Profit BRL'].sum()
total_profit_percent = df['Profit %'].sum()


# Cumulative profit over time
df['Date/Time'] = pd.to_datetime(df['Date/Time'])
df = df.sort_values('Date/Time')
cumulative_profit_brl = df['Cum. Profit BRL']

# Plotting cumulative profit trend
plt.figure(figsize=(12, 6))
plt.plot(df['Date/Time'], cumulative_profit_brl, label='Cumulative Profit BRL')
plt.xlabel('Date/Time')
plt.ylabel('Cumulative Profit (BRL)')
plt.title('Cumulative Profit Trend')
plt.legend()
plt.grid(True)
plt.show()

# Profit distribution
plt.figure(figsize=(12, 6))
plt.hist(df['Profit BRL'], bins=30, edgecolor='k', alpha=0.7)
plt.xlabel('Profit (BRL)')
plt.ylabel('Frequency')
plt.title('Distribution of Profits and Losses')
plt.grid(True)
plt.show()

desc_stats, trade_types_summary, signals_summary, total_profit_brl, total_profit_percent
```
